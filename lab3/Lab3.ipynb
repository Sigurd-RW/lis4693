```python
#Import Libraries

import pandas as pd
import altair as alt
from vega_datasets import data as vega_data
```

```python
#Load Library of Seattle Data

import requests
import io

url = "https://raw.githubusercontent.com/manika-lamba/SP26-LIS4_5693/refs/heads/main/lab-assignments/lab-3/Seattle-Library_2015-2021.csv"
response = requests.get(url)
response.raise_for_status() # Raise an exception for HTTP errors
text = response.text
```

```python
#Save as dataframe

seattle_df = pd.read_csv(io.StringIO(text))
seattle_df.head()
```

```python
#Group Materials
seattle_material_grouped = seattle_df.groupby(['MaterialType', 'CheckoutYear']).size().reset_index(name='counts')
```

```python
alt.Chart(seattle_material_grouped).mark_line(point=True).encode(
    x='CheckoutYear:O',
    y='counts:Q',
    color='MaterialType:N'
)
```

```python
selection = alt.selection_point(fields=['MaterialType'], bind='legend')

line_chart = alt.Chart(seattle_material_grouped).mark_line(point=True).encode(
    x='CheckoutYear:O',
    y='counts:Q',
    color='MaterialType:N',
    tooltip=['MaterialType', 'CheckoutYear', 'counts'],
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_params(
    selection
).interactive()
```
