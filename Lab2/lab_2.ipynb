```python
import spacy
```

## Load and read your file from GitHub


```python
import requests

url = "https://raw.githubusercontent.com/Sigurd-RW/lis4693/refs/heads/main/lab1/transcript.txt"
response = requests.get(url)
response.raise_for_status() # Raise an exception for HTTP errors
text = response.text
```

```python
print("Number of characters:", len(text)) # print number of characters

print(text[:100])   # print first 300 characters
```

## Sentence Segmentation

Create a blank spaCy model and add sentencizer.

```python
nlp = spacy.blank("en")
nlp.add_pipe("sentencizer")

doc = nlp(text)

sentences = list(doc.sents)

print("Number of sentences:", len(sentences))

print("\nFirst 5 sentences:")
for sent in sentences[:3]:
    print(sent)
```

## Word Count and Token Analysis

Let's count total words and unique words in your text file.

```python
words = [token.text.lower() for token in doc if token.is_alpha]

print("Total words:", len(words))
print("Unique words:", len(set(words)))
```

## Most Frequent Words

Let's find top 10 most frequent words in your file.

```python
from collections import Counter

word_freq = Counter(words)

print("Top 10 most frequent words:")
for word, count in word_freq.most_common(10):
    print(word, count)
```

All the most frequent words are gramatical syntax, which makes sense since they're the foundation of sentences.

## Using Full spaCy Pipeline

Now use the full model for better linguistic analysis.

```python
nlp2 = spacy.load("en_core_web_sm")

doc2 = nlp2(text)

print("Named Entities:")
for ent in doc2.ents:
    print(ent.text, ent.label_)
```

There are around 100 unique entities with about 13 types: PERSON,DATE,GPE,ORG,ORDINAL,CARDINAL,LOC,EVENT,WORK_OF_ART,PRODUCT,NORP,TIME, and PERCENT

## PhraseMatcher

```python
from spacy.matcher import PhraseMatcher

matcher = PhraseMatcher(nlp2.vocab, attr="LOWER")

phrases = ["display", "fold", "company"]

patterns = [nlp2(p) for p in phrases]

matcher.add("TECH_TERMS", patterns)

matches = matcher(doc2)

print("Matches found:")
for match_id, start, end in matches:
    print(doc2[start:end])
    print("Sentence:", doc2[start].sent)
```
